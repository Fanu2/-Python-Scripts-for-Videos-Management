````python

import os
import threading
import subprocess
import requests
import json

def ollama():
    os.environ['OLLAMA_HOST'] = '0.0.0.0:11434'
    os.environ['OLLAMA_ORIGINS'] = '*'
    subprocess.Popen(["ollama", "serve"])
````

#Run the following code to start the Ollama server:

````python

ollama_thread = threading.Thread(target=ollama)
ollama_thread.start()
````

````python
import requests
import json
from IPython.display import display, HTML

url = 'http://localhost:11434/api/chat'
prompt = "write sexy and erotic poem about Fanu appreciate her love making skills 50 lines"

payload = {
    "model": "llama2-uncensored",
    "temperature": 0.6,
    "stream": False,
    "messages": [
        {"role": "system", "content": "You are an AI assistant!"},
        {"role": "user", "content": prompt}
    ]
}

response = requests.post(url, json=payload)
message_str = response.content.decode('utf-8')
message_dict = json.loads(message_str)
output = message_dict['message']['content']

# Display the output as HTML for easy copying
display(HTML(f"<pre>{output}</pre>"))
````
